# Analyse prédictive du fonctionnement réflexif maternel : impact sur la qualité de la relation prénatale mère-enfant et les comportements de l'enfant #

Notre étude vise à mieux appréhender l’impact du fonctionnement réflexif de la mère  lors de sa grossesse   sur la qualité de la relation prénatale mère-enfant et la manière dont cela prédit les problèmes de comportement chez l’enfant après la naissance. 

Ainsi, par le biais des données collectées et des notions économétriques et statistiques, nous verrons si le niveau plus élevé de fonctionnement réflexif maternel pendant la grossesse sera positivement corrélé à une meilleure qualité de la relation prénatale mère-enfant. 

De plus, nous nous s’assurerons de vérifier s’il existe une relation significative entre la qualité de la relation prénatale mère-enfant et les problèmes de comportement de l'enfant.

**Objectif:**

Examiner comment le fonctionnement réflexif de la mère pendant la grossesse influence la qualité de la relation prénatale mère-enfant et comment cela prédit les problèmes de comportement chez l'enfant après la naissance. 

**Hypothèses:**

`H1 :` Un niveau plus élevé de fonctionnement réflexif maternel pendant la grossesse sera positivement corrélé à une meilleure qualité de la relation prénatale mère-enfant.

`H2 :` Un niveau plus élevé de fonctionnement réflexif chez la mère pendant la grossesse sera négativement corrélé à l'apparition de problèmes de comportement chez l'enfant.

``H3 : ``La qualité de la relation prénatale mère-enfant fonctionnera comme une variable médiatrice dans l'association entre le fonctionnement réflexif de la mère pendant la grossesse et la survenue de problèmes de comportement chez l'enfant.

``H4 :`` Il existe une relation significative entre la qualité de la relation prénatale mère-enfant et les problèmes de comportement de l'enfant


## Méthodologie de recherche

### Choix des variables et nature des natures des données

Notre analyse de recherche requiert des variables d’étude afin de pouvoir tester les hypothèses émises. Pour cela, nous avons opté pour des prédicteurs, variables cibles et enfin une variable médiatrice.

#### **Prédicteurs :**

- **``Fonctionnement réflexif maternel pendant la grossesse :``** Mesure la capacité de la mère à réfléchir sur ses propres états mentaux et ceux de son enfant à naître.

- **``Stress prénatal maternel :``**Évalue le niveau de stress ressenti par la mère pendant la grossesse.

- **``Support social pendant la grossesse :``** Évalue le niveau de soutien social perçu par la mère pendant la grossesse. Cela peut inclure le soutien du partenaire, de la famille, des amis, et des professionnels de santé.

#### **Variable cible :**

- **``Qualité de la relation mère-enfant et problèmes de comportement de l'enfant : ``** Une variable composite qui combine la qualité de la relation prénatale mère-enfant et les problèmes de comportement de l'enfant après la naissance. Par exemple, cela pourrait être une moyenne ou une somme des scores normalisés de ces deux aspects.

Notre étude de recherche vise à identifier les facteurs de risque potentiels pour les problèmes de comportement chez les enfants et à proposer des interventions pour soutenir les mères vulnérables pendant la grossesse.

### Description des tests d’hypothèses et de la régression linéaire

Nous utiliserons l’analyse de corrélation dans le but de décortique la liaison existante entre les prédicteurs et la variable cible. Précisément cette analyse mettra en avant les coefficients de corrélation pour tester nos hypothèses. Ensuite, nous enchaînerons avec une régression linéaire simple pour voir la relation entre le fonctionnement réflexif maternel pendant la grossesse et la qualité de la relation prénatale mère-enfant ainsi qu'entre le fonctionnement réflexif maternel et les problèmes de comportement chez l'enfant. 

Par la suite une régression multiple nous permettra de démystifier la relation entre le fonctionnement réflexif maternel, la qualité de la relation prénatale, et les problèmes de comportement en incluant la variable médiatrice. L’analyse de données se conclura sur les méthodes de médiation pour tester si la qualité de la relation prénatale mère-enfant médie l'effet du fonctionnement réflexif maternel sur les problèmes de comportement de l'enfant.

### Analyse de la médiation

```python
# analyse de la médiation

"Étape 1 : Vérifiez si X affecte Y (total effect)"
# Définir les variables
X = refletnorma[['Maternal Reflective Functioning', 'Prenatal Stress']] # les prédicteurs
Y = refletnorma['Composite Variable'] #variable cible 

# Ajoutez une constante (intercepte)
X = sm.add_constant(X)

# Modèle de régression
model1 = sm.OLS(Y, X).fit()
print(model1.summary())
```

```python
"Étape 2 : Vérifiez si X affecte M (path a)"
# Définir la variable médiatrice
M = refletnorma['Mother-Child Relationship Quality']

# Ajoutez une constante (intercepte)
X = sm.add_constant(X)

# Modèle de régression
model2 = sm.OLS(M, X).fit()
print(model2.summary())
```

```python
"Étape 3 : Vérifiez si M affecte Y et si X affecte Y en présence de M (direct effect"
X = refletnorma[['Maternal Reflective Functioning', 'Prenatal Stress','Mother-Child Relationship Quality']] # les prédicteurs
Y = refletnorma['Composite Variable'] #variable cible 

# Ajoutez une constante (intercepte)
X = sm.add_constant(X)

# Modèle de régression
model3 = sm.OLS(Y, X).fit()
print(model3.summary())
```
